```python
sns.set_style('darkgrid')
sns.countplot(y='Gender',hue='Branch',data=data,palette='RdBu', saturation=1)
```

```python
sns.set_style('darkgrid')
sns.countplot(y='Product line',hue='Branch',data=data,palette='RdBu_r', saturation=1)
```

```python
data_branch_product = data.groupby('Branch')['Product line'].value_counts()
print(data_branch_product)
data1 = data_branch_product.unstack()
data1.plot(kind='bar', figsize=(7, 5))
plt.ylabel('count')
plt.show()
```

```python
data_branch_gender = data.groupby('Branch')['Gender'].value_counts()
print(data_branch_gender)
data1 = data_branch_gender.unstack()
data1.plot(kind='bar', figsize=(7, 5))
plt.ylabel('count')
plt.show()
```
